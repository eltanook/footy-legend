```python
class Buscar(object):

    def jugador_partido(self, archivo='ruta_archivo_player'):
        '''recopila los datos del jugador relevantes para el partido (stamina, felicidad, habilidades) para que influyan en el juego
        archivo (diria que un csv xq es lo que se manejar) es un nombre placeholder, 
            debería buscar siempre el mismo (a no ser que queramos que se puedan guardar varios jugadores)
        esta solo se usaria una vez (cuando recien entras al juego) para buscar del csv y despues buscaria de los self
        pienso que se puede hacer de otra forma mas util pero para arrancar debiria andar bien algo de este estilo'''
        
        with open(archivo, 'rt') as f:
            datos = csv.DictReader(f)
            for fila in datos:
                self.fuerza = fila['fuerza']
                self.velocidad = fila['velocidad']
                self.resistencia = fila['resistencia']
                self.tecnica = fila['tecnica']
                self.vision = fila['vision']
                self.int_tactica = fila['inteligencia tactica']
                self.defensa = fila['defensa']
                self.toma_decisiones = fila['toma decisiones']
                self.teamwork = fila['teamwork']
                self.caracter = fila['caracter']
                self.resistencia_lesiones = fila['resistencia lesiones']
                self.mentalidad = fila['mentalidad']
                self.fisico = fila['fisico']
                self.liderazgo = fila['liderazgo']
            self.media = round(np.mean(self.fuerza, self.velocidad, self.resistencia, self.tecnica, self.vision, self.int_tactica, self.defensa, self.toma_decisiones, self.teamwork, self.caracter, self.resistencia_lesiones, self.mentalidad, self.fisico, self.liderazgo))
            self.age = datos['age'] #para el tema de que tan rapido baja la stamina
            self.stamina = datos['stamina'] 
        return self.media, self.age, self.stamina
                
    def resto_partido(lista_jugadores):
        '''busca los jugadores involucrados en el partido y los devuelve para que se balancee el partido acorde a las capacidades de los equipos
        lista_juagdores deberia tener solo los jugadores de la liga en la que juega el jugador o algo asi'''
        pass

    def actualizar_habilidad(self, habilidad_entrenada, entrenando=True): #entrenando es para chequear si es que el jugador entreno o si hay que bajarle stats xq no lo entrena hace mucho
        '''actualiza las habilidades del jugador desp de entrenamientos
        se puede implementar tambien que empiezen a bajar segun la edad
        :param habilidad_entrenada: int
        :return : int'''
        edad = self.age
        if entrenando:
            if 16<=edad<=24:
                habilidad_entrenada += 5 #estos valores habria que balancearlos, son placeholders
            elif 25<=edad<=32:
                habilidad_entrenada += 2
            else:
                habilidad_entrenada += 1
        else:
            if 16<=edad<=24:
                habilidad_entrenada -= 1 #estos valores habria que balancearlos, son placeholders
            elif 25<=edad<=32:
                habilidad_entrenada -= 2
            else:
                habilidad_entrenada -= 5
        return habilidad_entrenada

    def todos_los_jugadores(archivo='ruta_archivo_jugadores'):
        '''con pandas seguramente se pueda hacer mas optimo esto, es para tener una idea de los tiopos de datos con los que trabajan las otras funciones'''
        lista_jugadores = []
        with open(archivo, 'rt') as f:
            rows = csv.reader(f)
            headers = next(rows)
            for row in rows:
                record = dict(zip(headers, row))
                lista_jugadores.append(record)
        return lista_jugadores
    
    def jugadores_relevantes(self, lista_jugadores):
        '''busca juagdores que sirven para la sesion de juego y descarta el resto
        veremos como hacemos para implementar competencias internacionales mas adelante'''
        relevantes = []
        liga = self.liga
        for i in lista_jugadores:
            if i[liga] == self.liga:
                relevantes.append(i)
        return relevantes

    def obtener_nombre(self):
        nombre = ""
        input_box = pygame.Rect(100, 250, 200, 32)
        color_inactive = pygame.Color('lightskyblue3')
        color_active = pygame.Color('dodgerblue2')
        color = color_inactive
        active = False

        nombre_ingresado = False

        while not nombre_ingresado:
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    pygame.quit()
                    sys.exit()
                if event.type == pygame.MOUSEBUTTONDOWN:
                    if input_box.collidepoint(event.pos):
                        active = not active
                    else:
                        active = False
                    color = color_active if active else color_inactive
                if event.type == pygame.KEYDOWN:
                    if active:
                        if event.key == pygame.K_RETURN:
                            nombre_ingresado = True
                        elif event.key == pygame.K_BACKSPACE:
                            nombre = nombre[:-1]
                        else:
                            nombre += event.unicode

            self.screen.blit(self.fondo_cancha, (0, 0))

            txt_surface = self.font.render(nombre, True, color) #aca font con self? creo q si
            width = max(200, txt_surface.get_width() + 10)
            input_box.w = width
            self.screen.blit(txt_surface, (input_box.x + 5, input_box.y + 5))
            pygame.draw.rect(self.screen, color, input_box, 2)
            pygame.display.flip()

        return nombre
        
```

```python
import csv
import calendar

# Crear una lista para almacenar eventos
eventos = []

def mostrar_calendario(year):
    for month in range(1, 13):
        cal = calendar.month(year, month)
        print(f"Calendario para {calendar.month_name[month]} {year}:")
        print(cal)

def agregar_evento():
    year = int(input("Año del evento: "))
    month = int(input("Mes del evento: "))
    day = int(input("Día del evento: "))
    event = input("Descripción del evento: ")
    eventos.append((year, month, day, event))
    print(f"Evento agregado el {day}/{month}/{year}: {event}")

def guardar_eventos_en_archivo():
    with open('eventos.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["Año", "Mes", "Día", "Evento"])
        for evento in eventos:
            writer.writerow(evento)

def main():
    while True:
        print("1. Mostrar Calendario")
        print("2. Agregar Evento")
        print("3. Guardar Eventos (CSV)")
        print("4. Salir")
        
        choice = input("Elija una opción: ")
        
        if choice == '1':
            year = int(input("Año: "))
            mostrar_calendario(year)
        elif choice == '2':
            agregar_evento()
        elif choice == '3':
            guardar_eventos_en_archivo()
            print("Eventos guardados en 'eventos.csv'")
        elif choice == '4':
            break
        else:
            print("Opción no válida. Intente de nuevo.")

if __name__ == "__main__":
    main()
```
